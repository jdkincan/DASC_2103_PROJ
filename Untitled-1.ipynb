```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
df1 = df[["neighbourhood_group", "neighbourhood", "price", "number_of_reviews"]]
```

```python
df1_norm = df1.copy()
scaler = preprocessing.MinMaxScaler()
df1_norm[["price", "number_of_reviews"]] = scaler.fit_transform(df1_norm[["price", "number_of_reviews"]])
df1_norm = pd.get_dummies(df1_norm, columns=["neighbourhood_group", "neighbourhood"])
```

```python
df2 = df[['price', 'neighbourhood_group']]

df3 = df2.copy()

scaler = preprocessing.MinMaxScaler()
df3[["price"]] = scaler.fit_transform(df3[["price"]])
df3= pd.get_dummies(df3, columns=["neighbourhood_group"])
```

```python
df3.info()
```

```python
#Importing required modules
 
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
 
#Import KMeans module
from sklearn.cluster import KMeans
 
#Initialize the class object
kmeans = KMeans(n_clusters= 10)
 
#predict the labels of clusters.
label = kmeans.fit_predict(df3)
 
#Getting unique labels
u_labels = np.unique(label)
 
#plotting the results:
for i in u_labels:
    plt.scatter(df3[label == i , 0] , df3[label == i , 1] , label = i)
plt.legend()
plt.show()
```
